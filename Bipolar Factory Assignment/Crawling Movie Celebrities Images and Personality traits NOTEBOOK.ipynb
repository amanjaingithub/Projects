###### Fetching Celebrities name from **IDMB** website and storing it in a **Celebrity_Names.txt**

```python
# This for loop will request 2 pages of the site page 1 and page 2 and will fetch the names
for page in range(1,3):
    url = "https://www.imdb.com/list/ls002913270/?sort=list_order,asc&mode=detail&page="+str(page)
    request = rq.get(url)
    soup1 = BeautifulSoup(request.text,"html.parser")
    names = soup1.select(" h3 a")
    for name in names:
        with open("Celebrity_Names.txt","a") as f:
            f.write(str(name.text))
```


**personality_traits.txt :** This file contains all the possible personality traits which can be find in a celebrity such as - Romantic, Intelligent, Charming, Attractive, Confident, Disciplined and more.

The above code will fetch all the traits from the folder and store it in personality_traits variable

```python
personality_traits_file = open("personality_traits.txt", "r")
traits = personality_traits_file.read()
traits = traits.replace("\n"," ").split()
personality_traits = []
for trait in traits:
    trait = trait.lower()
    personality_traits.append(trait)
```

The above code will scrap the websites to get the information about celebrities and to get 10 images of every celebrity and will store those in the respective variable declared.Website which will be fetched are:

https://www.celebrities-galore.com/celebrities/      
https://www.gettyimages.in/

```python
celebrity_name =[]                  #This list will store all the names of celebrities
celebrity_personality_traits = []   #This list will store all the traits found in celebrities
celebrity_image = []                #This list will store the first 10 link found for the images for celebrity


# This function will return links for each celeb 
def find_celebrity_images(soup):   
    links = " "    #Initially it will be empty
    index = 1
    for item in soup.select(" figure>img"): #This will loop 10 times and concat the 10 links of images
        if item['src']:
            links = links + " " + item['src']
            if index == 10:
                break
            index +=1
        else:
            break
    return links


# This function will iterate through each word and will try to find if any trait is present and at the end 
# return the string contains every trait found with space seperated
def find_personality_traits(content):
    traits = " "
    for trait in personality_traits:
        if (trait in str(content).lower()) and (trait not in traits):
            traits = traits + trait + " " 
    return traits
    

    
# This function will simply return the Celebrity name
def get_celebrity_name(soup):
    name = soup.find(id = "profile-titles").find('h1').text
    return name



# This function will fetch the content about personalities of celebrities and pass those to find_personality_traits()
# function and return all the traits found
def get_celebrity_traits(soup_one, soup_two):
    content=""
    x = soup_one.select('div.page_content_padding_text p')
    y = soup_two.select('div.page_content_padding_text p')
    for data in x:
        content = content+""+data.text
    for data in y:
        content = content+" "+data.text
    traits = find_personality_traits(content)
    return traits
```

```python
# This loop will iterate for each personality
index = 0
for name in names:

    #url_images contains the url from where the images will be fetched from
    url_images = 'https://www.gettyimages.in/photos/'+str(name)+'?family=editorial&phrase='+str(name)+'&sort=mostpopular'

    #url_content_one and url_content_two contains the url from where data about personalities will be fetched
    url_content_one = "https://www.celebrities-galore.com/celebrities/"+str(CELEB_NAMES[index])+"/home/"
    url_content_two = "https://www.celebrities-galore.com/celebrities/"+str(CELEB_NAMES[index])+"/personality-number/"


    #These code will request and transform the webpages in the required form with the help of Requests and Beautiful Soup libs
    request_images = rq.get(url_images)   
    soup_images = BeautifulSoup(request_images.text, "html.parser") 

    request_content_one = rq.get(url_content_one)
    soup_content_one = BeautifulSoup(request_content_one.text, "html.parser")

    request_content_two = rq.get(url_content_two)
    soup_content_two = BeautifulSoup(request_content_two.text, "html.parser")

    celebrity_name.append(get_celebrity_name(soup_content_one)) #This will append the names in celebrity_names list
    celebrity_personality_traits.append(get_celebrity_traits(soup_content_one,soup_content_two)) #This will append the traits in celebrity_personality_traits list
    celebrity_image.append(find_celebrity_images(soup_images)) #This will append the links of image in celebrity_image list
    index = index + 1
```

```python
# This code will covert those lists into dataframe for further manipulations
data = {"Celebrity Name":celebrity_name,
        "Celebrity Images":celebrity_image,
        "Celebrity Personality Traits":celebrity_personality_traits}

df = pd.DataFrame(data)
df.head()
```

Now we need to split those 10 images links into different columns so it is easy to read and find

```python
# This function will help in spliting those links in 10 parts for each column
def split_images(list_of_images, number):
    if number:
        image = list_of_images.split()[number-1]
    return image
```

```python
# This loop will create 10 different columns for images
for i in range(1,11):
    df["Image "+str(i)] = df['Celebrity Images'].apply(split_images,number = i)
```

```python
# This will drop the image column which contains all the 10 links since we don't need it now
df.drop(["Celebrity Images"], axis = 1, inplace = True)
df.head()
```

Now we will create columns for traits which occured frequently in most of the celebrities and those which fans gets interested in

```python
#This function will return "Yes" if a trait is present in celebrity else "No"
def check_for_trait(data,trait):
    if trait in data:
        return "Yes"
    else:
        return "No"
```

```python
# This loop will create 6 different columns for different traits
for trait in ["Attractive","Romantic","Kind","Charming","Dignified","Wise"]:
    df[trait] = df['Celebrity Personality Traits'].apply(check_for_trait,trait=trait.lower())
```

```python
final_data = df[['Celebrity Name', 'Attractive', 'Romantic', 'Kind', 'Charming',
       'Dignified','Wise','Celebrity Personality Traits', 'Image 1', 'Image 2',
       'Image 3', 'Image 4', 'Image 5', 'Image 6', 'Image 7', 'Image 8',
       'Image 9', 'Image 10']]
final_data.rename(columns={'Celebrity Personality Traits':'More Traits'},inplace=True)
final_data.head()
```
